```python
import logging
from pathlib import Path

import numpy as np
import torch

from helpers import init_helper, data_helper, vsumm_helper, bbox_helper
from modules.model_zoo import get_model

from basic_NN.basic_NN import BasicRNN

import cv2
import matplotlib.pyplot as plt
```

```python
def read_frames(filename):
    video_cap = cv2.VideoCapture(filename)

    count = 0
    frames = []
    while True:
        success, frame = video_cap.read()

        if success:
            count += 1
            
            # convert from BGR to RGB format
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

            frames.append(frame)
        else:
            # print('Failed to read frame') # when last frame read
            break;


    print('Read ', count, ' frames in total')
    video_cap.release()
    
    return frames
```

```python
logger = logging.getLogger()

def evaluate(model, val_loader, nms_thresh, device):
    model.eval()
    stats = data_helper.AverageMeter('fscore', 'diversity')
    
    pred_sums = []
    test_keys = []
    user_sums = []

    with torch.no_grad():
        for test_key, seq, _, cps, n_frames, nfps, picks, user_summary in val_loader:

            # print(cps, n_frames, nfps, picks)

            seq_len = len(seq)
            seq_torch = torch.from_numpy(seq).unsqueeze(0).to(device)

            # Predicted Frame Importance Scores
            output = model.predict(seq_torch)

            pred_summ = vsumm_helper.get_keyshot_summ(
                output, cps, n_frames, nfps, picks)

            # print(pred_summ)

            eval_metric = 'avg' if 'tvsum' in test_key else 'max'
            # Key part here
            fscore = vsumm_helper.get_summ_f1score(
                pred_summ, user_summary, eval_metric)

            # print(list(pred_summ))
            # print(user_summary)

            pred_summ = vsumm_helper.downsample_summ(pred_summ)
            diversity = vsumm_helper.get_summ_diversity(pred_summ, seq)
            stats.update(fscore=fscore, diversity=diversity)
            
            user_summary = vsumm_helper.downsample_summ(user_summary[0])
            user_sum = np.array([False if x==0 else True for x in user_summary])
            pred_sums.append(pred_summ)
            test_keys.append(test_key)
            user_sums.append(user_sum)

    return stats.fscore, stats.diversity, pred_sums, test_keys, user_sums
```

```python
model_dir = "../models/ab_basic"
ckpt_path = "../models/ab_basic/checkpoint/summe.yml.0.pt"

model = BasicRNN(input_size=1024,output_size=1,hidden_dim=128,n_layers=3)
state_dict = torch.load(ckpt_path, map_location=lambda storage, loc: storage)
model.load_state_dict(state_dict)
```

```python
split_path = Path("../splits/summe.yml")
splits = data_helper.load_yaml(split_path)
stats = data_helper.AverageMeter('fscore', 'diversity')
nms_thresh = 0.4
device = "cuda:0"

model = model.eval().to(device)
```

```python
for split_idx, split in enumerate(splits):
    ckpt_path = data_helper.get_ckpt_path(model_dir, split_path, split_idx)
    state_dict = torch.load(str(ckpt_path),
                            map_location=lambda storage, loc: storage)
    model.load_state_dict(state_dict)

    val_set = data_helper.VideoDataset(split['test_keys'])
    val_loader = data_helper.DataLoader(val_set, shuffle=False)

    fscore, diversity, pred_sums, test_keys, user_sums = evaluate(model, val_loader, nms_thresh, device)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import groupby
from operator import itemgetter

def get_hlines_params(a):
    """Returns y, xmin, xmax arrays for plt.hlines."""
    data = []
    for y, row in enumerate(a):
        for b, col in groupby(enumerate(row), key=itemgetter(1)):
            if b == True:  # draw only True values
                x = list(col)
                xmin = x[0][0]
                xmax = x[-1][0] + 1
                data.append((y, xmin, xmax))
    return zip(*data)
```
